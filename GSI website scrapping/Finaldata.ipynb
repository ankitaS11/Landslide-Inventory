```python

# for manipulating dataframes
import pandas as pd
# for natural language processing: named entity recognition
import spacy
from collections import Counter
import en_core_web_sm
nlp = en_core_web_sm.load()
# for visualizations
%matplotlib inline
```

```python
tokens = nlp(''.join(str(final_data.Description.tolist())))
```

```python
items = [x.text for x in tokens.ents]
Counter(items)
```

```python
location_list = []
for ent in tokens.ents:
    if ent.label_ == 'GPE':
        location_list.append(ent.text)
        
location_counts = Counter(location_list).most_common(20)
df_location = pd.DataFrame(location_counts, columns =['Location', 'count'])
```

```python
df_location.plot.barh(x='Location', y='count', title="Landslide Prone Areas", figsize=(10,8)).invert_yaxis()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import geopandas as gpd
from urllib import request
from geotext import GeoText
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
from shapely.geometry import Point, Polygon
import descartes
```

```python
list_ = final_data.values.tolist()
list_
```

```python
cities = []
for x in list_:
  places = GeoText(str(x))
  city = list(places.cities)
  if city == []:
    cities.append('.')
  else:
    cities.append(city)
```

```python
cities
```

```python
df_city = pd.DataFrame(cities)
df_city
```

```python
"""Import Nominatim from geopy- Nominatim is a free service or tool or can be called as API with no keys that provides you with the data after providing it with name and address and vice versa.
On calling Nomination tool which accepts an argument of user_agent you can give any name as it considers it to be the name of the app to which it is providing its services. 
The geocode() function accepts the location name and returns a geodataframe that has all the details and since it’s a dataframe we can get the address, latitude and longitude by simply calling it with the given syntax """

"""if an individual search takes longer than  seconds, it’s not really a city name, and the script will thus weed out non-names from the final map. I put the city names and coordinates into a Pandas dataframe:"""
geolocator = Nominatim(timeout=5)
lat_lon = []
for city in cities: 
        location = geolocator.geocode(city)
        if location:
            print(location.latitude, location.longitude)
            lat_lon.append(location)
        else:
          lat_lon.append(['.','.'])
lat_lon
```

```python
"""if an individual search takes longer than  seconds, it’s not really a city name, and the script will thus weed out non-names from the final map. I put the city names and coordinates into a Pandas dataframe:"""
geolocator = Nominatim(timeout=5)
lat = []
lon = []
for city in cities: 
        location = geolocator.geocode(city)
        if location:
            print(location.latitude, location.longitude)
            lat.append(location.latitude)
            lon.append(location.longitude)
        else:
          lat.append(0)
          lon.append(0)
```

```python
df_lat = pd.DataFrame(lat)
df_lat.columns = ['Latitude']
df_lat
```

```python
df_lon = pd.DataFrame(lon)
df_lon.columns = ['Longitude']
df_lon
```

```python
df_latlong = pd.DataFrame(lat_lon)
df_latlong
```

```python
df_latlong.columns = ['Address', 'Latitude,Longitude']
```

```python
df_latlong
```

```python
df_latlong['Pin code'] = df_latlong['Address'].str.extract(r"\b(\d{6})\b")
df_latlong
```
